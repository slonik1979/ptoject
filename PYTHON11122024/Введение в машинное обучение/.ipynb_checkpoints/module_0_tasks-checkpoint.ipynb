```python
# Это ячейка 3.

def print_global_variable():
    """
    Производит печать значения переменной global_variable.
    """
    
    print(global_variable)
    
def add_suffix(value):
    """
    Изменяет переданную строку, присоединяя к ней справа случайное число.
    
    Аргументы:
        value: Строка, которую необходимо изменить.
        
    Возвращаемое значение:
        Исходная строка, к которой присоединено случайное число.
    """
    
    return value + ', к ней присоединили число ' + str(random.Random(0).randint(0, 10000))

def reinitialize_value():
    """
    Возвращает фиксированную строку.
    
    Возвращаемое значение:
        Строка «Изменённая глобальная переменная».
    """
    
    return 'Изменённая глобальная переменная'
```
